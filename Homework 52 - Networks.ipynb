#### Sayama 15.2:



_Calculate the following:_
- _The number of edges in a complete graph made of n nodes_

Each node is connected to $n-1$ nodes, and each connection happens only one way, so: $n(n-1)/2$

- _The number of edges in a regular graph made of n nodes each of which has
degree k_

Each node is connecter to $k$ nodes, and each connection happens only one way, so $nk/2$

#### Sayama 15.3:

_Draw the following graphs:_
- A complete graph made of four nodes
- A complete graph made of five nodes
- A bipartite graph made of a two-node group and a three-node group in which
all possible inter-group edges are present
- A bipartite graph made of two three-node groups in which all possible intergroup
edges are present

```
import networkx as nx
import matplotlib.pyplot as plt

g1 = nx.complete_graph(4)
g2 = nx.complete_graph(5)
g3 = nx.complete_bipartite_graph(2, 3)
g4 = nx.complete_bipartite_graph(3, 3)

plt.figure(figsize=(6,6))
plt.subplot(2,2,1)
nx.draw(g1)
plt.subplot(2,2,2)
nx.draw(g2)
plt.subplot(2,2,3)
nx.draw(g3)
plt.subplot(2,2,4)
nx.draw(g4)
```

#### Sayama 15.4:

_Every connected tree graph made of n nodes has exactly n-1 edges. Explain why._

Any node is connected to it's parent with exactly one edge. The only node that lacks that edge is the root node.

#### Sayama 15.10:

_Visualize the following graphs:_
documentation to learn how to generate them.
-  A “hypercube graph” of four dimensions.
- A “ladder graph” of length 5.
- A “barbell graph” made of two 20-node complete graphs that are connected
by a single edge.
- A “wheel graph” made of 100 nodes.

```
import networkx as nx
import matplotlib.pyplot as plt

g1=nx.hypercube_graph(4)
g2=nx.ladder_graph(5)
g3=nx.barbell_graph(20,0)
g4=nx.wheel_graph(100)

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
nx.draw(g1)
plt.subplot(2,2,2)
nx.draw(g2)
plt.subplot(2,2,3)
nx.draw(g3)
plt.subplot(2,2,4)
nx.draw(g4)
```

#### Excercise

_Imagine a graph representing hockey players and hockey clubs. Each node is either a player or a club and each edge indicates that a player has played for that club at some point in the past. Explain why this graph is a bipartite graph._

Clubs can't play for clubs, and players can't play for players. Necessarily an edge connects two nodes of different types.
